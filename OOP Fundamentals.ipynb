# ***Class***

```python
class Phone:
  def __init__ (self, model, battery, camera):
    self.model = model
    self.battery = battery
    self.camera = camera

  def charge(self, hour):
    pass

  def display_info(self):
    return f"Model: {self.model}, Battery: {self.battery}mAh, Camera: {self.camera}MP"

class MobilePhone:
  def __init__ (self, price):
    self.price = price
```

#***Inheritance***

```python
class SmartPhone(Phone):
  def __init__ (self, model, battery, camera, processor):
    super().__init__(model, battery, camera)
    self.processor = processor

  def display_info(self):
    return f"Model: {self.model}, Battery: {self.battery}mAh, Camera: {self.camera}MP, Processor: {self.processor}"
```

***Multiple Inheritance***

```python
class MyMobilePhone(MobilePhone, SmartPhone):
  def __init__ (self, model, battery, camera, processor, price):
    MobilePhone.__init__(self, price)
    SmartPhone.__init__(self, model, battery, camera, processor)

  def display_info(self):
    # Reusing SmartPhone's display_info and adding price from MobilePhone
    smartphone_info = SmartPhone.display_info(self)
    return f"{smartphone_info}, Price: {self.price}"
```

```python
my_mobile = MyMobilePhone("Redmi 10C", 5000, 50, "Snapdragon", 14000)

print(my_mobile.model) # from Phone class
print(my_mobile.battery) # from Phone class
print(my_mobile.camera) # from Phone class
print(my_mobile.processor) # from SmartPhone class
print(my_mobile.price) # from MobilePhone class
```

#***Polymorphism***

```python
class Camera:
  def __init__(self, name):
    self.name = name

  # method
  def take_picture(self):
    print(f"{self.name} is taking a picture.")
```

```python
class SmartMobilePhone(Camera):
  def __init__(self, name, resolution):
    super().__init__(name)
    self.resolution = resolution

  # method overriding
  def take_picture(self):
    print(f"{self.name} is taking a picture with a resolution of {self.resolution}MP.")

class DSLR(Camera):
  def __init__(self, name, resolution):
    super().__init__(name)
    self.resolution = resolution

  # method overriding
  def take_picture(self):
    print(f"{self.name} is taking a picture with a resolution of {self.resolution}MP.")

class Drone(Camera):
  def __init__(self, name, resolution):
    super().__init__(name)
    self.resolution = resolution

  # method override
  def take_picture(self):
    print(f"{self.name} is taking a picture with a resolution of {self.resolution}MP.")
```

```python
phone = SmartMobilePhone("Pixel 7Pro", 50)
dslr = DSLR("Nicon", 200)
drone = Drone("DJI", 100)

phone.take_picture()
dslr.take_picture()
drone.take_picture()
```

#***Encapsulation***

```python
class Mobile:
  def __init__(self, name, model, iemi):
    self.__name = name # private
    self.__model = model # private
    self.__iemi = iemi # private

  def name_getter(self):
    return self.__name

  def model_getter(self):
    return self.__model

  def iemi_getter(self):
    return self.__iemi

  def name_setter(self, new_name):
    self.__name = new_name

  def display_info(self):
    return f"Name: {self.__name}, Model: {self.__model}, IMEI: {self.__iemi}"
```

```python
myPhone = Mobile("iPhone", 17, "IPHONE123")
```

```python
# Access name and model using getter methods
print(f"Initial Name: {myPhone.name_getter()}")
print(f"Initial Model: {myPhone.model_getter()}")
print(f"Initial IMEI: {myPhone.iemi_getter()}")
```

```python
# Modify the name using the setter method
myPhone.name_setter("New iPhone Model")
print(f"Name after setter: {myPhone.name_getter()}")

# Display all information again to show the updated name and other private attributes
print(myPhone.display_info())
```

# ***Abstraction***

```python
from abc import ABC, abstractmethod

class Telephone(ABC):
  @abstractmethod
  def call(self):
    pass

  @abstractmethod
  def take_picture(self):
    pass
```

```python
class SPhone(Telephone):
  def call(self):
    print("Calling from Smart Phone")

  def take_picture(self):
    print("Taking picture from Smart Phone")

class IPhone(Telephone):
  def call(self):
    print("Calling from iPhone")

  def take_picture(self):
    print("Taking picture from iPhone")
```

```python
sPhone = SPhone()
sPhone.call()
sPhone.take_picture()

iPhone = IPhone()
iPhone.call()
iPhone.take_picture()
```
