```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic_data_set.csv')
```

```python
print(f"Total potential features: {df.shape[1]}")
print(f"Total sample data: {df.shape[0]}")
```

```python
df.sample(5)
```

```python
# missing values
df.isnull().sum()
```

```python
# duplicate values
df.duplicated().sum()
```

```python
# delete duplicate values
df.drop_duplicates(inplace=True)
```

```python
# information
df.info()
```

```python
# statistical values
df.describe()
```

***Univariate Analysis (Categorical Value)***

***Survived***

```python
sns.countplot(df, x=df['Survived'])
```

```python
df['Survived'].value_counts()
```

```python
survived_count = df['Survived'].value_counts()
survived_label = df['Survived'].unique()
plt.pie(survived_count, labels=survived_label, autopct="%.2f")
plt.title("Pie chart of Survival rate")
plt.show()
```

***Passenger Seat Class***

```python
sns.countplot(df, x=df['Pclass'])
```

```python
df['Pclass'].value_counts()
```

```python
survived_count = df['Pclass'].value_counts()
survived_label = df['Pclass'].unique()
plt.pie(survived_count, labels=survived_label, autopct="%.2f")
plt.title("Pie chart of Passenger Seat Class rate")
plt.show()
```

***Gender***

```python
sns.countplot(df, x=df['Sex'])
```

```python
df['Sex'].value_counts()
```

```python
survived_count = df['Sex'].value_counts()
survived_label = df['Sex'].unique()
plt.pie(survived_count, labels=survived_label, autopct="%.2f")
plt.title("Pie chart of Gender rate")

plt.show()
```

***Embarked***

```python
sns.countplot(df, x=df['Embarked'])
```

```python
df['Embarked'].value_counts()
```

```python
survived_count = df['Embarked'].value_counts()
survived_label = df['Embarked'].unique()
plt.pie(survived_count, labels=survived_count.index, autopct="%.2f")
plt.title("Pie chart of Embarked")

plt.show()
```

***Univariate Analysis (Numerical Column)***

***Age***

```python
sns.histplot(df['Age'], bins=50)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.show()
```

```python
sns.kdeplot(df['Age'])
plt.title('KDE Plot of Age')
plt.xlabel('Age')
plt.ylabel('Probability Distribution')

plt.show()
```

***Fare***

```python
sns.histplot(df['Fare'], bins=50)
plt.title('Histogram of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')

plt.show()
```

```python
sns.kdeplot(df['Fare'])
plt.title('KDE plot of Fare')
plt.xlabel('Fare')
plt.ylabel('Probability Distribution')

plt.show()
```
