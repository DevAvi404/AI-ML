```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic_data_set.csv')
```

***Gender wise Servival Rate***

```python
sns.countplot(x=df['Sex'], hue=df['Survived'])
```

```python
gender_split = df.groupby('Sex')['Survived'].value_counts()
print(gender_split)
```

```python
gender_split = df.groupby('Sex')['Survived'].value_counts(normalize=True) * 100
print(gender_split.map('{:.2f}%'.format))
```

***PClass wise Servival Rate***

```python
sns.countplot(x=df['Pclass'], hue=df['Survived'])
```

```python
pclass_split = df.groupby('Pclass')['Survived'].value_counts()
print(pclass_split)
```

```python
pclass_split = df.groupby('Pclass')['Survived'].value_counts(normalize=True) * 100
print(pclass_split.map('{:.2f}%'.format))
```
