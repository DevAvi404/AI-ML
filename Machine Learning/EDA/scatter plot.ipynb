```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df
```

```python
# total bill -> tips
sns.scatterplot(data=df,
                x='total_bill',
                y='tip')
```

```python
# gender, size => total bill -> tips
sns.scatterplot(data=df, 
                x='total_bill',
                y='tip', 
                hue='sex',
                size='size')
```
