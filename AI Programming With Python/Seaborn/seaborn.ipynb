```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
student = pd.read_csv('student_dataset.csv')
student.sample(5)
```

***Lineplot***

```python
# axes level
sns.lineplot(data=student, x='week', y='attendance_rate', hue='gender', errorbar=None)
```

```python
# figure level
sns.relplot(kind='line', data=student, x='week', y='attendance_rate', hue='class_level', errorbar=None)
```

***Scatter Plot***

```python
sns.scatterplot(data=student, x='study_hours', y='test_score')
```

```python
sns.scatterplot(data=student, x='study_hours', y='test_score', hue='gender', style='subject')
```

```python
sns.relplot(kind='scatter', data=student, x='study_hours', y='test_score', hue='gender', style='subject', size='Tshirt_size')
```

***Facet Plot***

```python
# figure level
sns.relplot(kind='scatter', data=student, x='study_hours', y='test_score', col='gender', style='subject')
```

```python
# figure level
sns.relplot(kind='scatter', data=student, x='study_hours', y='test_score', row='hostel', col='gender', style='subject')
```

```python
# figure level
sns.relplot(kind='scatter', data=student, x='study_hours', y='test_score', col='week', col_wrap=3)
```

***Histogram***

```python
# axes level
sns.histplot(data=student, x='attendance_rate', hue='gender', bins=15, element='step')
```

```python
# figure level
sns.displot(kind='hist', data=student, x='attendance_rate', hue='gender', bins=15, element='step')
```

```python
# figure level
sns.displot(kind='hist', data=student, x='attendance_rate', col='gender')
```

```python
# figure level
sns.displot(kind='hist', data=student, x='study_hours', col='gender')
```

***KDE Plot***

**kernel density estimation plot**

```python
# KDE plot
sns.kdeplot(data=student, x='attendance_rate', hue='gender', fill=True)
```

***Count Plot***

```python
sns.countplot(data=student, x='gender')
```

```python
sns.countplot(data=student, x='subject', hue='gender')
```

***Bar Plot***

```python
sns.barplot(data=student, x='gender', y='Marks', errorbar=None) # default 'mean'
```

```python
sns.barplot(data=student, x='gender', y='Marks', errorbar=None, estimator=np.max)
```

***Regression Plot***

```python
# axes level
sns.regplot(data=student, x='study_hours', y='test_score')
```

```python
# figure level
sns.lmplot(data=student, x='study_hours', y='test_score', hue='gender')
```

***Pair Plot***

```python
student_marks = student[['Marks', 'study_hours', 'attendance_rate', 'gender']]
student_marks.sample(3)
```

```python
sns.pairplot(data=student_marks)
```

```python
sns.pairplot(data=student_marks, hue='gender')
```

```python
sns.pairplot(data=student_marks, kind='hist', hue='gender')
```

***Joint Plot***

```python
sns.jointplot(data=student_marks, x='study_hours', y='Marks')
```

```python
sns.jointplot(data=student_marks, x='study_hours', y='Marks', hue='gender')
```

```python
sns.jointplot(data=student_marks, x='study_hours', y='Marks', kind='kde', hue='gender')
```
