***Data Load***

```python
import numpy as np

data = np.genfromtxt('student_scores.csv', delimiter=',', skip_header=True)
# column 1 -> Math marks, column 2 -> Science marks, column 3 -> English marks
print(f"Data Set:\n{data}") 
```

***Math***

```python
math_marks = data[ :: , :1]
print(f"Math Marks: {math_marks.T}")

# maximum number
max_math_marks = np.max(math_marks)
print(f"Maximum Math Marks: {max_math_marks}")

# minimum number
min_math_marks = np.min(math_marks)
print(f"Minimum Math Marks: {min_math_marks}")

# average
avg_math_marks = np.mean(math_marks)
print(f"Average Math Marks: {avg_math_marks}")

# meadian
median_math_marks = np.median(math_marks)
print(f"Median Math Marks: {median_math_marks}")

# standard deviation
std_math_marks = np.std(math_marks)
print(f"Standard Deviation: {std_math_marks}")
```

***Science***

```python
science_marks = data[ :: , 1:2]
print(f"Science Marks: {science_marks.T}")

# maximum number
max_science_marks = np.max(science_marks)
print(f"Maximum Science Marks: {max_science_marks}")

# minimum number
min_science_marks = np.min(science_marks)
print(f"Minimum Science Marks: {min_science_marks}")

# average
avg_science_marks = np.mean(science_marks)
print(f"Average Science Marks: {avg_science_marks}")

# meadian
median_science_marks = np.median(science_marks)
print(f"Median Science Marks: {median_science_marks}")

# standard deviation
std_science_marks = np.std(science_marks)
print(f"Standard Deviation: {std_science_marks}")
```

***English***

```python
english_marks = data[ :: , 2:3]
print(f"English Marks: {english_marks.T}")

# maximum number
max_english_marks = np.max(english_marks)
print(f"Maximum English Marks: {max_english_marks}")

# minimum number
min_english_marks = np.min(english_marks)
print(f"Minimum English Marks: {min_english_marks}")

# average
avg_english_marks = np.mean(english_marks)
print(f"Average English Marks: {avg_english_marks}")

# meadian
median_english_marks = np.median(english_marks)
print(f"Median English Marks: {median_english_marks}")

# standard deviation
std_english_marks = np.std(english_marks)
print(f"Standard Deviation: {std_english_marks}")
```

***Correlation Coefficient***

```python
study_hours = np.array([2, 4, 5, 7, 8])
exam_score = np.array([65, 75, 78, 88, 92])

data = np.array([study_hours, exam_score])

correlation = np.corrcoef(data)
print(f"Correlation Matrix:\n{correlation}")
```
