***ndArray***

```python
import numpy as np

#  1D Array

arr1 = np.array([1,2,3,4,5])
print(type(arr1))
```

```python
# 2D Array
arr2 = np.array([
    [1,2,3,4,5], # row 1
    [6,7,8,9,10] # row 2
])
print(type(arr2))
```

```python
# 3D Array
arr3 = np.array([
    # 1st floor
    [
        [1,2,3,4,5], # row 1
        [6,7,8,9,10] # row 2
    ],

    # 2nd floor
    [
        [1,2,3,4,5], # row 1
        [6,7,8,9,10] # row 2
    ],

    # 3rd floor
    [
        [1,2,3,4,5], # row 1
        [6,7,8,9,10] # row 2
    ]
])
print(type(arr3))
```

***Array Attributes***

```python
# dimenstion 
print(f"Dimention of Array 1: {arr1.ndim}")
print(f"Dimention of Array 2: {arr2.ndim}")
print(f"Dimention of Array 3: {arr3.ndim}")

print('\n')

# shape
print(f"Shape of Array 1: {arr1.shape}")
print(f"Shape of Array 2: {arr2.shape}")
print(f"Shape of Array 3: {arr3.shape}")

print('\n')

# data type
print(f"Data Type of Array 1: {arr1.dtype}")
print(f"Data Type of Array 2: {arr2.dtype}")
print(f"Data Type of Array 3: {arr3.dtype}")

print('\n')

# size
print(f"Size of Array 1: {arr1.size}")
print(f"Size of Array 2: {arr2.size}")
print(f"Size of Array 3: {arr3.size}")
```

***ndArray Data Type***

```python
arr =np.array([1,2,3])
print(arr.dtype)

arr = np.array([1.0, 2.0, 3.0])
print(arr.dtype)

# upcasted to float
arr = np.array([1,2,3.0])
print(arr.dtype) 

# upcasted to string
arr = np.array([1,2.0,'3'])
print(arr.dtype)
```

***Selecting a Data Type for an Array***

```python
arr = np.array([1,2,3], dtype=np.int8) # range: -128 to +128
print(arr.dtype)

arr = arr.astype(np.int64)
print(arr.dtype)
```

```python
# from dictionary
dic = {
    'a' : 10,
    'b' : 20,
    'c' : 30,
    'd' : 40,
    'e' : 50
}

keys = dic.keys()
values = dic.values()
items = dic.items()

# keys to nd array
arr = np.array(list(keys))

print(f"Array: {arr}")
print(f"Type: {type(arr)}")
print(f"Data Type: {arr.dtype}")
print(f"Shape: {arr.shape}")
print(f"Dimention: {arr.ndim}")
print(f"Size: {arr.size}")

print('\n')

# values to nd array
arr = np.array(list(values))

print(f"Array: {arr}")
print(f"Type: {type(arr)}")
print(f"Data Type: {arr.dtype}")
print(f"Shape: {arr.shape}")
print(f"Dimention: {arr.ndim}")
print(f"Size: {arr.size}")

print('\n')

# items to nd array
arr = np.array(list(items))

print(f"Array: {arr}")
print(f"Type: {type(arr)}")
print(f"Data Type: {arr.dtype}")
print(f"Shape: {arr.shape}")
print(f"Dimention: {arr.ndim}")
print(f"Size: {arr.size}")
```

***Indexing & Slicing***

```python
# indexing
print(arr1) # 1D array
print(arr1[2])
arr1[2] = 10
print(arr1)

print("\n")

print(arr2) #2D array
print(arr2[0][1])
arr2[0][1] = 10
print(arr2)
```

```python
# slicing
# 1D array -> arr[start : end : step]
arr1_mod = arr1[1:4] # by default 1 sstep
print(arr1_mod)

print("\n")

# 2D array -> arr[row_start : row_end : step , column_start : column_end : step]
print(arr2[0:1, ]) # get 0th row
print(arr2[1: , ]) # get 1st row

print("\n")

print(arr2[ : , 0:1]) # get 0th column
print(arr2[ : , 1:2]) # get 1st column
```

```python
# slicing
# arr1_mod in a view of arr1 [not copy]
print(arr1)
print(arr1_mod)

arr1_mod[0] = 1000 # also change in main array

print(arr1)

print("\n")

arr1_mod = arr1[2:4].copy() 
print(arr1)
print(arr1_mod)

arr1_mod[0] = 500 # only change the sliced array

print(arr1_mod)
print(arr1) # not change in main array
```

***Advanced Indexing***

```python
# 1D array
lst = np.array([10, 20, 30, 40, 50])
values = lst[[0,3,4]]
print(values)

print("\n")

# 2D array
print(arr2)
values = arr2[[0,1], [1,2]]
print(values)

print("\n")

# by boolean 
print(arr2[arr2<2]) # print less than 2 values at arr2
```
