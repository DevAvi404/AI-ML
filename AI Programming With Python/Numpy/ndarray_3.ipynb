```python
import numpy as np

arr = np.array([3, 5, 2, 5, 6, 9, 8])
```

***Sorting***

```python
# implace
sorted_arr = arr.copy()
sorted_arr.sort()

print(f"Sorted Array: {sorted_arr}")
```

```python
# copy sort
sorted_arr = np.sort(arr)

print(f"Sorted Array: {sorted_arr}")
```

***2D Array Sorting***

```python
matrix = np.array([
    [2, 4, 9, 5, 8], 
    [5, 3, 2, 8, 7]])

# horizontal axis = 1
hor_sort = np.sort(matrix, axis=1)
print(f"Sorted Array (Horizontally):\n{hor_sort}")

print('\n')

# vertical axis = 0
hor_sort = np.sort(matrix, axis=0)
print(f"Sorted Array (Vertically):\n{hor_sort}")
```

***Searching***

```python
# np.where(condition, x, y) if true x, else replace with y
index = np.where(arr > 5)
print(index)

index = np.where(arr > 5, arr, 0) # returns array
print(index)
```

***2D Array***

```python
index = np.where(matrix > 5)
print(index)

index = np.where(matrix > 5, matrix, 0)
print(index)
```

```python
# maximum value
maximim_value_index = np.argmax(arr)
print(f"Maximim Value Index: {maximim_value_index}")

# minimum value
minimum_value_index = np.argmin(arr)
print(f"Minimum Value Index: { minimum_value_index}")
```

***2D Array***

```python
# maximum value
maximim_value_index = tuple(map(int, np.unravel_index(np.argmax(matrix), matrix.shape)))
print(f"Maximim Value Index: {maximim_value_index}")

# minimum value
minimum_value_index = tuple(map(int, np.unravel_index(np.argmin(matrix), matrix.shape)))
print(f"Minimum Value Index: { minimum_value_index}")
```

***Counting***

```python
arr = np.random.randint(1, 100, size=100)
print(f"Array: {arr}")

x = np.count_nonzero(arr > 60)
print(f"Value Greater than 60: {x}")
```

***Unique Value***

```python
# unique value
arr = np.random.randint(1, 10, size=10)
print(f"Array: {arr}")

unique_value = np.unique(arr)
print(f"Unique Value: {unique_value}")


unique_value = np.unique(arr, return_counts=True)
print(f"Unique Values with Mapping: {unique_value}")
```
