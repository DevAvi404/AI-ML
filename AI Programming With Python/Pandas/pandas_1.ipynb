***Transforming to Data Frame from Inbuild Data***

```python
# from list
myList = [
    ['A', 25],
    ['B', 30],
    ['C', 28]
]

listDF = pd.DataFrame(myList, columns=['Name', 'Age'], index=[1, 2, 3])

print(f"Type: {type(listDF)}")
listDF
```

```python
# from tuple
myTuple = (
    ('A', 25),
    ('B', 30),
    ('C', 28)
)

tupleDF = pd.DataFrame(myTuple, columns=['Name', 'Age'], index=[1, 2, 3])

print(f"Type: {type(tupleDF)}")
tupleDF
```

```python
# from dictionary 
myDict = {
    'Name' : ['A', 'B', 'C'],
    'Age' : [25, 30, 28]
}

dictDF = pd.DataFrame(myDict, index=[1, 2, 3])

print(f"Type: {type(dictDF)}")
dictDF
```

```python
# from dictionary 
myDict = [
    {'Name' : 'A', 'Age' : 25, 'City' : 'Chattogram'},
    {'Name' : 'B', 'Age' : 30, 'City' : 'Dhaka'},
    {'Name' : 'C', 'Age' : 28}
]

dictDF = pd.DataFrame(myDict, index=[1, 2, 3])

print(f"Type: {type(dictDF)}")
dictDF
```

***Changing Values and Iteration***

```python
# row deletion
data_frame
```

```python
data_frame.drop(0)
```

```python
# column deletion
data_frame.drop('Instructor', axis=1)
```

```python
# updating value
data_frame.loc[1, 'Python Marks'] = 90
data_frame.loc[1, 'CompletionStatus'] = 'Completed'
data_frame.loc[1]
```

```python
# iteration
# slower
for i, series in data_frame.iterrows():
    print(f"{i} : {series}")
```

```python
# faster
for i in data_frame.itertuples(index=False):
    print(i)
```
