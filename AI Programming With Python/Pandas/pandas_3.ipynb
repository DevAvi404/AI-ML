***Apply Function***

```python
# min max scaling

mn = df['Total Marks'].min()
mx = df['Total Marks'].max()

df['Scaled Marks'] = df['Total Marks'].apply(lambda x : (x - mn) / (mx - mn))

df.sample(3)
```

```python
# custom build function

def grading_system(marks):
    if marks >= 260:
        return 'A+'
    elif marks >= 210:
        return 'A'
    else:
        return 'A-'

df['Grade'] = df['Total Marks'].apply(grading_system)

df.sample(3)
```

***Date-Time***

```python
d_f = pd.read_csv('student_completed_data.csv')

d_f.sample(3)
```

```python
# data type: object -> datetime
d_f['EnrollmentDate'] = pd.to_datetime(d_f['EnrollmentDate'])
d_f['FinishedDate'] = pd.to_datetime(d_f['FinishedDate'])
```

```python
d_f['Enrollment Day'] = d_f['EnrollmentDate'].dt.day_name()
d_f['Enrollment Month'] = d_f['EnrollmentDate'].dt.month_name()
d_f['Enrollment Year'] = d_f['EnrollmentDate'].dt.year
d_f['Total Time'] = d_f['FinishedDate'] - d_f['EnrollmentDate']
d_f.sample(3)
```

***Groupby***

```python
d_f = pd.read_csv('student_completed_data.csv')

d_f.sample(3)
```

```python
instructor_group = d_f.groupby('Instructor')
instructor_group
```

```python
for instructor, df_ins in instructor_group:
    print(f"\nInstructor: {instructor}")
    display(df_ins)
```

```python
instructor_group.describe()
```

```python
mean_marks = instructor_group['Total Marks'].mean()
mean_marks
```

```python
# replacing null value with respect to groupby (instructor) mean
d_f['Total Marks'] = d_f['Total Marks'].fillna(d_f['Instructor'].map(mean_marks))
d_f.sample(3)
```
