```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

***2D Line Plot***

```python
df = pd.read_csv('enrollment_data.csv')
df.sample(3)
```

```python
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title('Student enrollment over the years')

plt.plot(df['Year'], df['Programming'], label='Programming')
plt.plot(df['Year'], df['Digital Marketing'], label='Digital Marketing')
plt.legend() # for labeling
plt.show()
```

```python
# adding some visualization
plt.xlabel('Years')
plt.ylabel('Count of Enrollment')
plt.title('Student enrollment over the years')

plt.plot(df['Year'], df['Programming'], label='Programming', color='red', linewidth=2, linestyle='dashed', marker='o', markersize=5)
plt.plot(df['Year'], df['Digital Marketing'], label='Digital Marketing', color='green')
plt.legend() # for labeling
plt.grid() # for gird
plt.show()
```

***Scatter Plot***

```python
student = pd.read_csv('student_IQdata.csv')
student.sample(3)
```

```python
# Relation Between Study Hour & IQ Score (Positive Correlation)
plt.xlabel('Study Hour')
plt.ylabel('IQ Score')
plt.title('Relation Between Study Hour & IQ Score')

plt.scatter(student['Study_Hour'], student['IQ_Score'])
plt.show()
```

```python
# Relation Between Chilling Hours & IQ Score (Negative Correlation)
plt.xlabel('Chilling Hours')
plt.ylabel('IQ Score')
plt.title('Relation Between Chilling Hours & IQ Score')

plt.scatter(student['Chilling_Hours'], student['IQ_Score'])
plt.show()
```

```python
# Relation Between Shoe Size & IQ Score (No Correlation)
plt.xlabel('Study Hour')
plt.ylabel('Shoe Size')
plt.title('Relation Between Shoe Size & IQ Score')

plt.scatter(student['Shoe_Size'], student['IQ_Score'])
plt.show()
```

***Bar Chart***

```python
data = {
    'Student': [ 'B', 'C', 'D', 'E', 'F', 'G', 'H'],
    'Gender': [ 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male'],
    'StudyHours': [ 2, 5, 3, 7, 4, 8, 1],
    'Attendance': [ 60, 88, 70, 95, 80, 98, 55],
    'Grade': [ 'Fail', 'Pass', 'Fail', 'Pass', 'Pass', 'Pass', 'Fail']
}

df = pd.DataFrame(data)
df
```

```python
# gender wise 
gender_group = df.groupby('Gender').size()
gender_group
```

```python
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Bar Chart of Gegnder among Student')

plt.bar(gender_group.index, gender_group.values, color=['orange', 'blue'], edgecolor='black')
plt.show()
```

```python
# bar chart on real data
df = pd.read_csv('student_data.csv')
df.sample(3)
```

```python
# Completion Status
status = df.groupby('CompletionStatus').size()
status
```

```python
plt.xlabel('Completion Status')
plt.ylabel('Count')
plt.title('Bar Chart of Completion Status among Student')

plt.bar(status.index, status.values, color=['green', 'blue', 'red'], edgecolor='black')
plt.show()
```

```python
# location
location = df.groupby('Location').size()
location
```

```python
plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Bar Chart of Location among Student')

plt.bar(location.index, location.values, color=['yellow', 'blue', 'red', 'green', 'orange'], edgecolor='black')
plt.show()
```

***Pie Chart***

```python
plt.title('Pie Chart of Completion Status among Student')
plt.pie(status, labels=status.index, autopct='%1.1f%%', shadow=True)
plt.show()
```
