```python
import numpy as np

arr = np.random.randint(1, 100, size=(10, 5)) # sample = 10, feature = 5
print(arr)
```

```python
# reshape (row->column, column->row)
new_arr = arr.reshape(5,10)
print(new_arr)
```

```python
# 2D -> 1D
oneD_array = new_arr.flatten() # row wise
print(oneD_array)

oneD_array = np.ravel(new_arr, order='F') # column wise
print(oneD_array)
```

```python
a = np.random.randint(1,10, size=(2,3))
b = np.random.randint(20,30, size=(2,3))

print(f"Array a:\n{a}")
print('\n')
print(f"Array b:\n{b}")
```

***Concatenate***

```python
con_row = np.concatenate((a,b), axis=0) # row wise (a's row = b's column)
con_col = np.concatenate((a,b), axis=1) # column wise (a's column = b'row)

print(f"Concatenated Array (Row Wise):\n{con_row}")
print('\n')
print(f"Concatenated Array (Column Wise):\n{con_col}")
```

***Transpose***

```python
matrix = np.random.randint(1,10, size=(2,3))
print(f"Matrix:\n{matrix}")

print('\n')

transpose_matrix = matrix.T
print(f"Transpose Matrix:\n{transpose_matrix}")
```

***Split***

```python
arr = np.random.randint(1,10, size=(10,))
print(f"Array: {arr}")

print('\n')

splited_arr = np.array_split(arr, 3)
print(f"Splited Array: {splited_arr}")

# equal division: np.split(arr, size) note: len(arr) must be divisible by the size
```

***Arethmetic Operators & Mathematical Functions***

```python
x = np.array([10, 20, 30, 40, 50])
y = np.array([1, 2, 3, 4, 5])
```

***Operators***

```python
add = x + y # x[i] + y[i]
print(f"Addition: {add}")

sub = x - y # x[i] - y[i]
print(f"Subraction: {sub}")

mul = x * y # x[i] * y[i]
print(f"Multiplication: {mul}")

div = x / y # x[i] / y[i]
print(f"Division: {div}")

rem = x % y # x[i] % y[i]
print(f"Remainder: {rem}")
```

***Functions***

```python
# trigonometry
# sin, cos, tan takes values in radian 
sin_val = np.sin(x)
print(f"sin(0): {sin_val}")

cos_val = np.cos(x)
print(f"cos(0): {cos_val}")

tan_val = np.tan(x)
print(f"tan(0): {tan_val}")

deg_rad = np.deg2rad(x) # 0 * 3.1416 / 180
print(f"Degree to Radian: {deg_rad}")

rad_deg = np.rad2deg(deg_rad)
print(f"Radian to Degree: {rad_deg}")
```

```python
# logarithmic
base_10 = np.log10(x)
print(f"10 Based Logarithmic Value: {base_10}")

sqrt = np.sqrt(x)
print(f"Square Root Value: {sqrt}")
```

***Logical Functions***

```python
# comparison: >, <, >=, <=, ==, !=
greater_than = x > y
print(f"Is Greater Than? -> {greater_than}")

equal = x == y
print(f"Is Equal? -> {equal}")

# all() -> is  all true?
# any() -> is any true?
print(f"Is All Ture? -> {np.all(greater_than)}")
print(f"Is Any Ture? -> {np.all(equal)}")
```
