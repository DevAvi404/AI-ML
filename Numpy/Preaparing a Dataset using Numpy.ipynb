```python
import numpy as np

# dataset
data = np.array([
    [9.5, 0.5, 95.5], # 0st index
    [8.2, 1.2, 86.4], # 1st index
    [7.8, 1.5, 82.1], # 2nd index
    [7.0, 1.0, 81.0], # 3rd index
    [6.5, 2.0, 74.5], # 5th index
    [5.0, 1.5, 70.0], # 6th index
    [4.5, 3.0, 60.5], # 7th index
    [3.0, 2.5, 55.0], # 8th index
    [2.0, 4.0, 44.0], # 9th index
    [1.5, 4.5, 39.5] # 10th index
])
```

***Randomizing the Dataset***

```python
n = len(data)

random_index = np.random.choice(n, size=(n,), replace=False)

data = data[random_index]
```

***Features & Target Extraction***

```python
# features
x = data[ : , : 2] # row-> all, column-> 0st, 1st

# target
y = data[ : , 2 : ] # row-> ll, column-> 2nd
```

***Split***

```python
# 80% data -> train, 20% data -> test

train_index = int(len(data) * 0.8)

# training data
x_train = x[ : train_index, : ]
y_train = y[ : train_index, : ]

# testing data
x_test = x[train_index : , : ]
y_test = y[train_index : , : ]
```
